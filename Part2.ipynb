## 2.1 Baseline í‰ê°€ì§€í‘œë¥¼ ê³„ì‚°í•˜ëŠ” `calculate_baseline_score` í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

í‰ê·  baselineì˜ $MAE$ (Mean Absolute Error)ì™€ $R^2$ ì ìˆ˜ë¥¼ ê³„ì‚°í•˜ëŠ” `calculate_baseline_score` í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”(ë³„ë„ë¡œ í•™ìŠµê³¼ í‰ê°€ ë°ì´í„°ë¥¼ ë‚˜ëˆ„ì§€ ì•Šì•„ë„ ë©ë‹ˆë‹¤).

````python
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np


def calculate_baseline_score(homes):
    mae = float('inf')
    r2 = -1

    # YOUR CODE HERE

    target = 'SalePrice'

    y_train = homes[target]
    predict = y_train.mean()
    y_pred = [predict] * len(y_train)

    mae = mean_absolute_error(y_train, y_pred)
    r2 = r2_score(y_train, y_pred)

    # YOUR CODE ENDS HERE
    
    return mae, r2
````

````python
mae, r2 = calculate_baseline_score(homes)
print(f"MAE: {mae:.2f}, R^2: {r2:.2f}")
````

## 2.2 ë°ì´í„°ì…‹ì„ ì•„ë˜ì˜ ê¸°ì¤€ì„ ì‚¬ìš©í•´ì„œ train/validation/test ì…‹ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” `split_home_data` í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

- **Train**: 2006ë…„ë¶€í„° 2008ë…„ê¹Œì§€ íŒ”ë¦° ì£¼íƒë“¤ (1,920ê°œ)

- **Validation**: 2009ë…„ì— íŒ”ë¦° ì£¼íƒë“¤. (644ê°œ)

- **Test**: 2010ë…„ì— íŒ”ë¦° ì£¼íƒë“¤. (340ê°œ)

````python
def split_home_data(homes):
    train = None
    validation = None
    test = None

    # YOUR CODE HERE
    
    train = homes.loc[(homes['YrSold'] >= 2006) & (homes['YrSold'] <= 2008)]

    validation = homes.loc[homes['YrSold'] == 2009]

    test = homes.loc[homes['YrSold'] == 2010]
    
    # YOUR CODE ENDS HERE
    
    return train, validation, test
````

````python
train, validation, test = split_home_data(homes.copy())
````

````python
train.shape, validation.shape, test.shape
````

## 2.3 train / validation / test ì…‹ì„ features(X)ì™€ target(y)ìœ¼ë¡œ ê°ê° ë‚˜ëˆ„ëŠ” `divide_home_data` í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

- íŠ¹ì„±ì—ëŠ” ìµœì†Œ ë‘ ê°œì˜ ìˆ˜ì¹˜í˜• (numeric) íŠ¹ì„±ê³¼ ë²”ì£¼í˜• (categorical) íŠ¹ì„±ì„ í¬í•¨í•˜ì„¸ìš”. 
- ìœ„ì˜ ì¡°ê±´ì„ ë§Œì¡±í•œë‹¤ë©´ ì¶”ê°€ featureë¥¼ ì„ íƒí•˜ëŠ” ê²ƒì— ëŒ€í•œ ì œí•œì€ ì—†ìŠµë‹ˆë‹¤.

````python
def divide_home_data(data):
    X = None
    y = None

    # YOUR CODE HERE
    
    target = 'SalePrice'

    X = data.drop(columns=target)
    y = data[target]
    
    # YOUR CODE ENDS HERE

    return X, y
````

````python
X_train, y_train = divide_home_data(train)
X_val, y_val = divide_home_data(validation)
X_test, y_test = divide_home_data(test)
````

## 2.4 One-hot encodingì„ ì¹´í…Œê³ ë¦¬í˜• íŠ¹ì„±(feature)ì— ìˆ˜í–‰í•˜ëŠ” `encoding_one_hot` í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”

````python
from category_encoders import OneHotEncoder


def encoding_one_hot(X_train, X_val, X_test):
    encoder = None  # OneHotEncoder ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•˜ì„¸ìš”
    X_train_encoded = None
    X_val_encoded = None
    X_test_encoded = None

    # YOUR CODE HERE
    
    encoder = OneHotEncoder(use_cat_names = True)
    X_train_encoded = encoder.fit_transform(X_train)
    X_val_encoded = encoder.transform(X_val)
    X_test_encoded = encoder.transform(X_test)
    
    # YOUR CODE ENDS HERE
    
    return encoder, X_train_encoded, X_val_encoded, X_test_encoded
````

````python
encoder, X_train_encoded, X_val_encoded, X_test_encoded = encoding_one_hot(X_train, X_val, X_test)
````

## 2.5 Ridge regression ëª¨ë¸ì„ í•™ìŠµ (fit)í•˜ê³  í•™ìŠµëœ ëª¨ë¸ê³¼ validation score(MAE, $R^2$)ì„ ë°˜í™˜í•˜ëŠ” `fit_ridge_regression`í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

````python
import numpy as np
from sklearn.linear_model import RidgeCV

def fit_ridge_regression(X_train_encoded, X_val_encoded, y_train, y_val):
    ridge = RidgeCV(alphas=np.arange(0.01, 2, 0.01), normalize=True, cv=5)
    mae = float('inf')
    r2 = -1

    # YOUR CODE HERE
    
    ridge.fit(X_train_encoded, y_train)

    y_pred = ridge.predict(X_train_encoded)   

    mae = mean_absolute_error(y_train, y_pred)
    r2 = r2_score(y_train, y_pred)
    
    # YOUR CODE ENDS HERE
    
    return ridge, mae, r2
````

````python
ridge, mae, r2 = fit_ridge_regression(X_train_encoded, X_val_encoded, y_train, y_val)
print(f'alpha: {ridge.alpha_}')
print(f'cv best score: {ridge.best_score_}')
print("Ridge Regression ê²€ì¦ë°ì´í„° MAE: ", mae)
````
